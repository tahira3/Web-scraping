# Import necessary libraries

```python
import requests
from bs4 import BeautifulSoup as bs 
import re
```

# Load the webpage

```python
r = requests.get('https://www.basketball-reference.com/leagues/NBA_2023_per_game.html')
soup = bs(r.content)
# soup
```

# Start scraping data from the webpage

```python
table = soup.select('table.sortable')[0]
column = table.find('tr').find_all('th')
column_name = [c.string for c in column]
column_name
```

```python
tbody = table.find('tbody').select('tr.full_table')
tbody
l=[]
for tr in tbody:
    t_data = tr.find_all('td')
    data = [t.get_text() for t in t_data]
    l.append(data)
l[0]    
```

# Load data into Pandas DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(l,columns = column_name[1:] )
df
```
