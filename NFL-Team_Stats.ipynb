# Load in the necessary libraries

```python
import requests
from bs4 import BeautifulSoup as bs 
import re
```

# Load our webpage

```python
r = requests.get("https://www.nfl.com/stats/team-stats/")
```

```python
soup = bs(r.content)
```

```python
# print(soup.prettify())
```

```python
table = soup.select('table.d3-o-table')[0]
# print(get_table.prettify())
```

# Scrape all the column name from the table

```python
thead = table.find('thead').find_all('th')
column_header = [c.get_text() for c in thead]
column_header
```

# Scrape all the rows from the table

```python
tbody = table.find('tbody').find_all('tr')
list = []
for tr in tbody:
    td = tr.find_all('td')
    data =[td.get_text() for td in td]
    list.append(data)
list[0]
```

# Load the table into a DataFrame

```python
import pandas as pd
```

```python
df = pd.DataFrame(list,columns = column_header )
df.head(3)
```

```python
tbody = table.find('tbody').find_all('tr')
def get_team_name(name):
    if name.find('div'):
        return [team.get_text() for team in name.select('div.d3-o-club-fullname')]
    else:
        return name.get_text()
    
list = []
for tr in tbody:
    td = tr.find_all('td')
    data =[get_team_name(td) for td in td]
    list.append(data)
# list

df = pd.DataFrame(list,columns = column_header)
df
```
