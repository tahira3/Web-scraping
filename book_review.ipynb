Import Necessary Libraries

```python
import requests
from bs4 import BeautifulSoup as bs 
import re
import pandas as pd
import numpy as np
```

Start Scraping

```python
pages = np.arange(1,51)
title = []
price = []
stock = []
rating =[]


for page in pages:
    url = "https://books.toscrape.com/catalogue/page-"+str(page)+".html"
    r = requests.get(url)
    soup = bs(r.content)
#     print(soup.prettify())
#     print(url)

    for t in soup.find_all('h3'):
        title.append(t.a['title'])
    
    
    for p in soup.select('p.price_color'):
        price.append(p.get_text().replace('£',''))
        
    
    for s in soup.select('p.instock'):
        stock.append(s.get_text().strip())
    
    
    for r in soup.select('article'):
        rating.append(r.p['class'][1])
        
# title
```

Load Data into a Pandas DataFrame

```python
data = {'Title':title,'Ratings':rating,'Price':price,'Availability':stock}
df = pd.DataFrame(data)
df
```
